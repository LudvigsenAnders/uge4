```python
def generate_numbers(max_value):
    num = 1
    while num <= max_value:
        yield num
        num += 1

# Using the generator
for value in generate_numbers(5):
    print(value)
```

```python
# Generator expression to calculate squares
squares = (x * x for x in range(1, 6))
for square in squares:
    print(square)
```

```python
def fibonacci_sequence(limit):
    a, b = 0, 1
    while a < limit:
        yield a
        a, b = b, a + b

# Using the generator
for num in fibonacci_sequence(10):
    print(num)
```

```python
def is_palindrome(num):
    # Skip single-digit inputs
    if num // 10 == 0:
        return False
    temp = num
    reversed_num = 0

    while temp != 0:
        reversed_num = (reversed_num * 10) + (temp % 10)
        temp = temp // 10

    if num == reversed_num:
        return num
    else:
        return False
```

```python
def infinite_palindromes():
    num = 0
    while True:
        if is_palindrome(num):
            i = (yield num)
            if i is not None:
                num = i
        num += 1
```

```python
pal_gen = infinite_palindromes()
for i in pal_gen:
    digits = len(str(i))
    pal_gen.send(10 ** (digits))
```

```python
pal_gen = infinite_palindromes()
for i in pal_gen:
    print(i)
    digits = len(str(i))
    if digits == 5:
        pal_gen.throw(ValueError("We don't like large palindromes"))
    pal_gen.send(10 ** (digits))
```

```python
# initialize generators "again" to reset them for reuse sine they are exhausted after one pass

lines = (line for line in open(file_name))
list_line = (s.rstrip().split(",") for s in lines)
cols = next(list_line)
housing_dicts = (dict(zip(cols, data)) for data in list_line)
```

```python
def counter(start=0):
    """Generator that counts items."""
    num = start
    while True:
        yield num
        num += 1
```

```python
# Initialize counter
villa_counter = counter(start=1)
```

```python

purchase_price = (
    float(housing_dict["purchase_price"])
    for housing_dict in housing_dicts
    if housing_dict["house_type"] == "Villa"
    and next(villa_counter)  # Count each villa
)

```

```python
total_villa_purchaseAmt = 0
total_villa_count = 0

for price in purchase_price:
    total_villa_purchaseAmt += price
    total_villa_count += 1
```

```python
avg_villa_price = total_villa_purchaseAmt / total_villa_count if total_villa_count > 0 else 0
print(f"Average villa purchase price: DKK {avg_villa_price}")
```
