```python
from pathlib import Path


# Current file's directory
cwd = Path.cwd()

path = cwd.parent.parent

file_path = path/'data'/'Navneliste.txt'
```

```python
print(file_path)
```

```python
with open(file_path, encoding="utf-8") as f:
    read_data = f.read()


```

```python
names = [ name.strip() for name in read_data.split(',')]
```

```python
names_sorted_by_char = sorted(names, key=str.lower)
```

```python
names_sorted_by_len = sorted(names, key=len)
```

```python
my_char_dict = {}
```

```python
for name in names:
    for char in name:
        char = char.lower()
        if char not in my_char_dict:
            my_char_dict[char] = 1
        else:
            my_char_dict[char] += 1
```
